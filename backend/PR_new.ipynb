```python
from sklearn.preprocessing import LabelEncoder

# List of columns with object dtype (assumed categorical)
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Label encode categorical columns
label_encoder = LabelEncoder()
for col in categorical_cols:
    if col in df.columns:
        df[col] = label_encoder.fit_transform(df[col].astype(str))

# Find correlation matrix
correlation_matrix = df.corr()

# Get highest correlated columns
highest_corr_cols = correlation_matrix.abs().nlargest(5, 'collision_severity')['collision_severity'].index.tolist()
# Replace 'target_column' with the column you want to find correlations with

# Display highest correlated columns
print("Highest Correlated Columns:")
print(highest_corr_cols)
```
